```python
# cell 01 Octal to Decimal

#I edited code from Dr. Biersach for this

def octal_to_decimal(o_str):
    # Reverse string 'o' because the
    # LSB must be in first position
    o_str = o_str[::-1]
    sum = 0
    for position in range(len(o_str)):
        weight = pow(8, position)
        digit = int(o_str[position])
        sum += digit * weight
    return sum


o_str = "2323"
d = octal_to_decimal(o_str) #look nice large numbr
print(f"{o_str} in octal = {d:,} in decimal")
```

```python
#I edited code from Dr. Biersach for this

# Cell 02 - Decimal to Octal

def decimal_to_octal(d):
    o_str = ""
    while d > 0:
        # Append the octal digit to the string
        o_str += str(d % 8)
        d = int(d / 8)
    # Reverse string 'o' because the
    # LSB must be in first position
    o_str = o_str[::-1]
    return o_str


d = 1235
o_str = decimal_to_octal(d)
print(f"{d:,} in decimal = {o_str} in binary")
```

```python
#I edited code from Gemini AI for this

# cell 03

import numpy as np

def octal_to_binary_info(octal_str_input: str) -> dict:
    """
    Converts an octal string to its binary representation and provides related information.

    Args:
        octal_str_input (str): The octal number as a string (e.g., "123", "0o123").

    Returns:
        dict: A dictionary containing:
            - 'input_octal_string': The original octal string input.
            - 'decimal_value': The integer decimal value of the octal string.
            - 'binary_string': The binary string representation (e.g., '1010011').
            - 'binary_array': A list of 0s and 1s representing the binary string.
            - 'powers_of_two_context_for_binary': A list of powers of two corresponding
                                                 to the positions in the binary_array (MSB to LSB).
    """
    # Validate input: Ensure it's a string and looks like an octal number
    # (can start with '0o' or just be digits 0-7).
    if not isinstance(octal_str_input, str):
        raise TypeError("Input must be a string representing an octal number.")

    # Remove '0o' prefix if present for robust parsing
    cleaned_octal_str = octal_str_input.lower()
    if cleaned_octal_str.startswith('0o'):
        cleaned_octal_str = cleaned_octal_str[2:]

    # Check if the cleaned string contains only valid octal digits (0-7)
    if not all(char in '01234567' for char in cleaned_octal_str):
        raise ValueError(f"Invalid octal digits found in '{octal_str_input}'. "
                         "Octal numbers must only contain digits 0-7.")

    # Handle empty string or string '0' as valid octal representing 0
    if not cleaned_octal_str or cleaned_octal_str == '0':
        return {
            'input_octal_string': octal_str_input,
            'decimal_value': 0,
            'binary_string': '0',
            'binary_array': [0],
            'powers_of_two_context_for_binary': [1] # 2^0
        }

    # Convert octal string to decimal integer
    decimal_value = int(cleaned_octal_str, 8)

    # Convert decimal integer to binary string
    binary_str = bin(decimal_value)[2:] # Slice [2:] to remove '0b' prefix

    # Create the binary array from the binary string
    binary_array = [int(bit) for bit in binary_str]

    # Determine the powers of two context for the binary representation
    num_binary_bits = len(binary_str)
    powers_of_two_context = []
    for i in range(num_binary_bits):
        powers_of_two_context.append(2**(num_binary_bits - 1 - i))

    return {
        'input_octal_string': octal_str_input,
        'decimal_value': decimal_value,
        'binary_string': binary_str,
        'binary_array': binary_array,
        'powers_of_two_context_for_binary': powers_of_two_context
    }

# --- Example Usage ---
if __name__ == "__main__":
    # --- Example Usage for Octal to Binary Conversion ---
    print("\n" + "#"*60)
    print("Octal to Binary Conversion Examples:")
    print("#"*60)

    octal_numbers = ["2323"]

    for oct_num_str in octal_numbers:
        try:
            binary_info = octal_to_binary_info(oct_num_str)
            print(f"\n--- Processing Octal: '{binary_info['input_octal_string']}' ---")
            print(f"  Decimal Value: {binary_info['decimal_value']}")
        except (ValueError, TypeError) as e:
            print(f"Error for octal input '{oct_num_str}': {e}")

    # Example of invalid octal input
    try:
        octal_to_binary_info("0o89")
    except (ValueError, TypeError) as e:
        print(f"\nError test for '0o89': {e}")

    try:
        octal_to_binary_info(123) # Incorrect type
    except (ValueError, TypeError) as e:
        print(f"\nError test for 123 (integer): {e}")

    print("\n" + "#"*60)
    print("End of Octal to Binary Conversion Example.")
    print("#"*60)

```

```python
#I edited code from Gemini AI for this

# cell 04

import numpy as np

def decimal_to_octal_info(decimal_num_input: int) -> dict:
    """
    Converts a decimal integer to its octal string representation and provides related information.

    Args:
        decimal_num_input (int): The non-negative decimal integer to convert.

    Returns:
        dict: A dictionary containing:
            - 'input_decimal_number': The original decimal integer input.
            - 'octal_string': The octal string representation (e.g., '173').
            - 'octal_array': A list of integers representing each digit of the octal string.
            - 'powers_of_eight_context_for_octal': A list of powers of eight corresponding
                                                    to the positions in the octal_array (MSB to LSB).
    """
    # Validate input: Ensure it's a non-negative integer.
    if not isinstance(decimal_num_input, int) or decimal_num_input < 0:
        raise ValueError("Input 'decimal_num_input' must be a non-negative integer.")

    # Handle the special case for 0
    if decimal_num_input == 0:
        return {
            'input_decimal_number': 0,
            'octal_string': '0',
            'octal_array': [0],
            'powers_of_eight_context_for_octal': [1] # 8^0
        }

    # Convert decimal integer to octal string.
    # oct(decimal_num_input) returns a string like '0o173'. We slice [2:] to remove the '0o' prefix.
    octal_str = oct(decimal_num_input)[2:]

    # Create the 'octal_array' by converting each character digit to an integer.
    # This array represents each digit in the octal number.
    octal_array = [int(digit) for digit in octal_str]

    # Determine the 'powers_of_eight_context_for_octal' list.
    # The length of the octal string tells us the number of octal digits/positions.
    # For example, if octal_str is '173' (length 3), powers are 8^2, 8^1, 8^0.
    num_octal_digits = len(octal_str)
    powers_of_eight_context = []
    for i in range(num_octal_digits):
        # Calculate 8 raised to the power of (num_octal_digits - 1 - i)
        # This correctly generates powers from MSB down to LSB.
        powers_of_eight_context.append(8**(num_octal_digits - 1 - i))

    return {
        'input_decimal_number': decimal_num_input,
        'octal_string': octal_str,
        'octal_array': octal_array,
        'powers_of_eight_context_for_octal': powers_of_eight_context
    }


# --- Example Usage ---
if __name__ == "__main__":
    print("\n" + "#"*60)
    print("Decimal to Octal Conversion Examples:")
    print("#"*60)

    decimal_number = [1235]

    for dec_num in decimal_number:
        try:
            octal_info = decimal_to_octal_info(dec_num)
            print(f"\n--- Processing Decimal: '{octal_info['input_decimal_number']}' ---")
            print(f"  Octal String: {octal_info['octal_string']}")

            print(f"  Powers of Eight Context (for Octal): {octal_info['powers_of_eight_context_for_octal']}")
        except (ValueError, TypeError) as e:
            print(f"Error for decimal input '{dec_num}': {e}")

    print("\n" + "#"*60)
    print("End of Decimal to Octal Conversion Examples.")
    print("#"*60)


```
