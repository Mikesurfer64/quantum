```python
"""quantum_circuit2.ipynb"""

# Cell 01 - Circuit to produce only two states (63 or 64 decimal)
# Reference: Gemini AI to find combo of X and CNOT gates

from IPython.display import display
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import (
    plot_distribution,
)
from qiskit_aer import AerSimulator

qc = QuantumCircuit(7)

qc.save_statevector("sv1")
qc.h(6)
qc.cx(6, 0)
qc.save_statevector("sv2")
qc.x(0)
qc.cx(6, 1)
qc.save_statevector("sv3")
qc.x(1)
qc.cx(6, 2)
qc.save_statevector("sv4")
qc.x(2)
qc.cx(6, 3)
qc.save_statevector("sv5")
qc.x(3)
qc.cx(6, 4)
qc.save_statevector("sv6")
qc.x(4)
qc.cx(6, 5)
qc.save_statevector("sv7")
qc.x(5)

qc.measure_all()
# Do not modify the following code
display(qc.draw(output="mpl"))

backend = AerSimulator()
qc_transpiled = transpile(qc, backend)
result = backend.run(qc_transpiled, shots=100_000).result()

counts = result.get_counts(qc)
display(plot_distribution(counts))
```
