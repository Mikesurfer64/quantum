```python
import numpy as np
import pandas as pd

# Cell 01 - Extract the petal dimensions from the Iris dataset
# Iris CSV taken from https://archive.ics.uci.edu/dataset/53/iris

file_name = "iris.csv"
petal_data = np.genfromtxt(
    file_name, delimiter=",", skip_header=1, usecols=(2, 3), dtype=float
) #datatype as floats, usecols 2d matrix with 2 columns

pd.DataFrame(#create dataframe 
    petal_data,
    columns=["Petal Length (cm)", "Petal Width (cm)"], #text headers 
)
```

```python
# Cell 02 - Extract the species from the Iris dataset

species_data = np.genfromtxt(
    file_name, delimiter=",", skip_header=1, usecols=(4), dtype=str
)

pd.DataFrame(species_data, columns=["Species"]) #extract one column, daratype=string
```

```python
import matplotlib.pyplot as plt

# Cell 03 - Plot the relationship between petal length and petal width between varieties

plt.figure(figsize=(8, 6))
for species in np.unique(species_data): #mask becomes an array of booleans (T or F)
    mask = species_data == species  # Broadcasting and element-wise comparison, does it equal? w/ use of ==
    plt.scatter(petal_data[mask, 0], petal_data[mask, 1], label=species.strip('"')) #Each T in masks Boolean index array tells Numpy to include the corresponding element from the petal_data array
plt.title("Iris Petal Length vs Petal Width By Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(loc="upper left")
plt.show()
```
